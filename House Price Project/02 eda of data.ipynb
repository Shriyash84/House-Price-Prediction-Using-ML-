```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd. read_csv("cleaned_data.csv")
```

```python
ds.info()
```

```python
# POSTED_BY - pie chart represent percentage
# UNDER_CONSTRUCTION - pie chart represent percentage
# RERA - pie char represent percentage
#  BHK_NO - pie char represent percentage
# READY_TO_MOVE - pie chart represent percentage
# RESALE - pie chart represent percentage
# CITY - Bar charts
```

```python
posted_by = ds.groupby("POSTED_BY")["POSTED_BY"].count()
```

```python
posted_by.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6))
```

```python
under_construction = ds.groupby("UNDER_CONSTRUCTION")["UNDER_CONSTRUCTION"].count()
under_construction 
```

```python
under_construction.plot(kind='pie', autopct='%1.1f%%')
```

```python
rera = ds.groupby("RERA")["RERA"].count()
rera
```

```python
rera.plot(kind='pie', autopct='%1.1f%%')
```

```python
bhk = ds.groupby("BHK_NO.")["BHK_NO."].count()
bhk.plot(kind='bar', figsize=(10, 6), color='skyblue')
# Adding labels and title
plt.xlabel('BHK Number')
plt.ylabel('Count')
plt.title('Count of BHK Numbers')
```

```python
citywisecount = ds.groupby("CITY")["CITY"].count()
citywisecount = citywisecount.sort_values(ascending=False)
counts = citywisecount.head(10)
```

```python
ax = counts.plot(kind='bar',color='skyblue',figsize=(8,6))
ax.legend(["Count of Entries"],loc="upper right")
# Set Label of Titles 
plt.xlabel("Number of BHK")
plt.ylabel("Count of Entries")
```
